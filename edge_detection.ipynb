```python
img_1=cv2.imread('07-junk.jpg');
```

```python
img_2=cv2.imread("Millennium_Bridge.jpg")
```

```python
img_3=cv2.imread("roadjpg.jpg")
```

```python
# Convert the image to grayscale
gray_img1=cv2.cvtColor(img_1,cv2.COLOR_BGR2GRAY)
gray_img2=cv2.cvtColor(img_2,cv2.COLOR_BGR2GRAY)
gray_img3=cv2.cvtColor(img_3,cv2.COLOR_BGR2GRAY)
```

```python
# Show the grayscale images
plt.imshow(gray_img1,cmap='gray')
plt.show()
```

```python
plt.imshow(gray_img2,cmap='gray')
plt.show()
```

```python
plt.imshow(gray_img3,cmap='gray')
plt.show()
```

```python
# Perform the edge detection using cv2.canny function
edges_img1=cv2.Canny(gray_img1,threshold1=80,threshold2=150)
plt.imshow(edges_img1, cmap='gray');
```

```python
edges_img2=cv2.Canny(gray_img2,threshold1=100,threshold2=200)
plt.imshow(edges_img2,cmap='gray');
```

```python
edges_img2=cv2.Canny(gray_img3,threshold1=100,threshold2=200)
plt.imshow(edges_img2,cmap='gray');
```
